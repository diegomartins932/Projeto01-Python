```python
#Realizando importação da biblioteca pandas
import pandas as pd
```

```python
#Carregando arquivo csv
df = pd.read_csv("Dados/vendas.csv", parse_dates=['data'], dayfirst=True)
```

```python
#Mostrando as 5 primeiras linhas do dataset
df.head()
```

```python
#Mostrando as últimas 5 linhas do dataset
df.tail()
```

```python
#Mostrando os tipos de dados do dataset
df.dtypes
```

```python
#Realizando operações com data
#Mostrando meses
df.data.dt.month
```

```python
#Mostrando ano
df.data.dt.year
```

```python
#definindo uma nova coluna
df['Mes'] = df.data.dt.month
```

```python
#Mostrando a nova coluna
df.head()
```

```python
#Realizando filtros - Mês
df.loc[df.Mes == 12]
```

```python
#Realizando filtros - produto
df.loc[df.produto == 'prod_1']
```

```python
#Filtro com condição - Clausula And
filtro_1 = df.produto == 'prod_1'
filtro_2 = df.pais == 'Brasil'
df.loc[(filtro_1) & (filtro_2)]
```

```python
#Filtro com condição - Clausula Or
filtro_1 = df.Mes == 6
filtro_2 = df.pais == 'Alemanha'
df.loc[(filtro_1) | (filtro_2)]
```

```python
#Filtro por intervalo
filtro_int = (df.Mes > 7) & (df.Mes <=12)
df.loc[filtro_int]
```

```python
#filtro de analise por conteudo - semelhante ao like '%....'
filtro = df.pais.str[0] == 'A'
df.loc[filtro]
```

```python
#filtro de analise por conteudo - semelhante ao like '....%'
filtro = df.pais.str[-1] == 'l'
df.loc[filtro]
```

```python
#filtro de analise por conteudo - semelhante ao like '%...%'
filtro = df.pais.str.contains('s')
df.loc[filtro]
```

```python
#filtro de analise por conteudo - semelhante ao like '%....' com or
filtro = df.pais.str.contains('si|Al')
df.loc[filtro]
```

```python
#Criando sub dataset
df2010 = df.loc[df.data.dt.year == 2010]
```

```python
#Mostrando 10 linhas
df2010.head(10)
```

```python
#Utilizando clasula group by - sum
df2010.groupby(['produto']).quantidade.sum()
```

```python
#Utilizando clausula group by - size
df2010.groupby(['produto']).size().sort_values()
```

```python
#Utilizando clausula group by - size (decrescente)
df2010.groupby(['produto']).size().sort_values(ascending=False)
```

```python
#Retornando a menor valor venda
df['valor_total'].min()
```

```python
#Retornando a maior valor venda
df['valor_total'].max()
```

```python
#Retornando a média de venda
#Retornando a menor valor venda
df['valor_total'].mean()
```

```python
#Trabalhando com gráficos
#Gráfico de barras
df["produto"].value_counts(ascending=False).plot.bar();
```

```python
#Gráfico de barras horizontais
df["pais"].value_counts().plot.barh();
```

```python
#Gráfico de barras horizontais - decrescente
df["pais"].value_counts(ascending=True).plot.barh();
```

```python
#Gráfico de Pizza
df.groupby(df["data"].dt.year)["valor_total"].sum().plot.pie();
```

```python
#Adicionando um título e alterando o nome dos eixos
import matplotlib.pyplot as plt
df["pais"].value_counts().plot.bar(title="Total vendas por País")
plt.xlabel("pais")
plt.ylabel("Total Vendas");
```

```python
#Alterando a cor
df["pais"].value_counts().plot.bar(title="Total vendas por Cidade", color="orange")
plt.xlabel("País")
plt.ylabel("Total Vendas");
```
